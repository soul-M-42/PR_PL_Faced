```python
import numpy as np
import pickle
result_dir = './result/Faced/Faced_Result.pkl'
with open(result_dir, 'rb') as file:
    result = pickle.load(file)
    acc = result['best_acc_mat']
```

```python
acc
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 示例数据
categories = []
values = acc.tolist()
values = sorted(values)
plt.figure(figsize=(14, 8))

# 创建柱状图
plt.bar(range(123), values)

# 添加红色水平虚线
y_value = 1.0 / 9  # 水平虚线的位置
plt.axhline(y=y_value, color='red', linestyle='--', linewidth=2)
plt.axhline(y=np.mean(values), color='blue', linestyle='--', linewidth=2)

# 添加标题和标签
plt.title('Faced 9-class cross-subject accuracy PR_PL')
plt.xlabel('')
plt.ylabel('Top1 acc')

# 显示图表
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


# 将值与原始索引配对，并排序
values_with_indices = sorted(enumerate(acc), key=lambda x: x[1])
indices, values = zip(*values_with_indices)

# 创建图表并指定宽高
plt.figure(figsize=(14, 8))

# 创建柱状图
plt.bar(indices, values)

# 添加红色水平虚线
y_value = 1.0 / 9  # 水平虚线的位置
plt.axhline(y=y_value, color='red', linestyle='--', linewidth=2)

# 添加标题和标签
plt.title('Bar Chart with Horizontal Line')
plt.xlabel('Original Index')
plt.ylabel('Values')

# 显示图表
plt.show()
```
